```python
from mpl_toolkits import mplot3d 
%matplotlib inline 
import numpy as np 
import matplotlib.pyplot as plt

def f(x, y): 
    return (x*y)

x = np.linspace(-6, 6, 30) 
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y) 
Z = f(X, Y)

fig = plt.figure() 
ax = plt.axes(projection='3d') 
ax.contour3D(X, Y, Z, 50, cmap='binary') 
ax.set_xlabel('x') 
ax.set_ylabel('y') 
ax.set_zlabel('z')
```

```python
from mpl_toolkits.mplot3d import axes3d
u = np.linspace(1.5*np.pi, -1.5*np.pi, 100) 
# This returns a sequence of evently spaced numbers starting at 1.5*np.pi and ending at -1.5*np.pi with a sample size of 100.

[x,y] = np.meshgrid(u, u) 
# This returns a 3-D coordinate matrix based on the values from u, creating the grid that can be seen in the figure below

squared = np.sqrt(x.flatten()**2 + y.flatten()**2) 
z = np.cos(squared) 
# This returns the non-negative square root of squared, which is a list of the sums of the squares of corresponding x and y values
z = z.reshape(x.shape) 
# Changes the shape of the array z to an array of size x.shape which is (100,100)

fig = plt.figure() 
ax = fig.add_subplot(111, projection='3d') 
# Creates a new subplot to add to the figure fig with a 3d projection
ax.plot_wireframe(x, y, z, rstride=10, cstride=10)
# Plots a 3D wireframe with the dimensions of x, y, and z, along with an array row step size of 10 and column step size of 10
ax.view_init(elev=50., azim=30)
# Sets the elevation(the angle of elevation) to 50 and azimuth(angle in the x-y plane) of the axes to 30
plt.savefig("figure1.png")
```
